```python
page = 1
total_users = []
headers = {
    "x-api-key": "reqres-free-v1",
    "Accept" : "application/json"
}
while True:
    url = f"https://reqres.in/api/users?page={page}"
    a = requests.get(url=url, headers=headers)
    if a.status_code == 404:
        print("404 error")
        break
    a.raise_for_status()
    users = a.json().get("data",[])
    print(users)
    if not users:
        break
    page += 1
    total_users.extend(users)
len(total_users)
```

```python
def extract_top_nav_urls(base_url):
    resp = requests.get(base_url)
    resp.raise_for_status()

    soup = BeautifulSoup(resp.text, "lxml")
    # CSS selector for the top nav bar links
    nav_links = soup.select("ul.navigation.menu li a")  # <ul class="navigation menu"> … <a href="…">
    
    # Build absolute URLs and dedupe
    urls = {urljoin(base_url, a["href"]) for a in nav_links}

    return sorted(urls)

if __name__ == "__main__":
    base = "https://www.python.org"
    urls = extract_top_nav_urls(base)
    print("Top navigation URLs on python.org:")
    for u in urls:
        print(u)
```

```python
with requests.get("https://httpbin.org/image/png", stream = True) as website:
    website.raise_for_status()
    with open("image.png", "wb") as out_file:
        for chunk in website.iter_content(1024):
            if not chunk:
                continue
            print("1KB Done")
            out_file.write(chunk)
```

```python
import time
requestcount = 0
waittime = 1
while requestcount != 5:
    website = requests.get("https://httpbin.org/status/429")
    if website.status_code != 429:
        print("connection worked")
        break
    print(f'{waittime} {requestcount}')
    time.sleep(waittime)
    requestcount += 1
    waittime *= 2
```
